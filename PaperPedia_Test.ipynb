# City Category Pie Chart

```python
dataset1['City'] = pd.factorize(dataset1.City_Category)[0]
dataset1.head()
```

    pie() function generates pie graph representing data in an array which we convert from string type to int type in the above
    command and used it in groupby for analysing it with string data for pie graph.
    axis is a function of the matplotlib library that adds axis to the current graph and makes it as current axis.

```python
sums = dataset1.groupby(dataset1["City_Category"])["City"].sum()
axis('equal');
pie(sums,labels=sums.index);
show()
```
